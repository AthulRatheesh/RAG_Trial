```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader,service_context
from llama_index.llms.huggingface import HuggingFaceLLM #to interact with HF LLM
from llama_index.core.prompts.prompts import SimpleInputPrompt

model_config = {}
model_config['protected_namespaces'] = ()
```

```python
documents=SimpleDirectoryReader("/content/data").load_data()
documents[0]
```

```python
system_prompt='''
You are a Q&A assistant for the course of Artificial Intelligence for KTU S7
students of the CSE program.
Your goal is to answer questions as accurately as possible based on the
context provided.
You should only provide hyperlinks that reference the context below. Do NOT
include links to other sites. If the context is empty or you are unsure of the
answer, you should say "I do not know." Your answers should be verbose and detailed.
'''
#For Llama2
#Default format supported by Llama2
query_wrapper_prompt=SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch  #To import LLama2

llm=HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature":0.0, "do_sample":False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="meta-llama/Llama-2-7b-chat-hf",
    model_name="meta-llama/Llama-2-7b-chat-hf",
    device_map="auto",
    model_kwargs={"torch_dtype":torch.float16, "load_in_8bit":True}
)
```

```python
!pip install langchain_huggingface
%pip install llama-index-embeddings-langchain
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model=LangchainEmbedding(HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))  #search the model
```

```python
#Now indexing and querying using ServiceContext(deprecated)...bundles them into one..now Settings
from llama_index.core import Settings
'''service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)'''
Settings.chunk_size=1024
Settings.llm=llm
Settings.embed_model=embed_model
Settings.system_prompt=system_prompt
Settings.query_wrapper_prompt=query_wrapper_prompt
```

```python
index=VectorStoreIndex.from_documents(documents,embed_model=embed_model)
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("How are you?")
print(response)
```
